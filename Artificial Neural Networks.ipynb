## Activation Functions

Activation functions in neural networks are crucial because they introduce non-linear properties to the network, enabling it to learn more complex relationships in the data. Hereâ€™s an overview of the main types of activation functions, their common uses, and Python code examples using TensorFlow/Keras, which is one of the most popular libraries for deep learning.

### 1. Sigmoid:

**Formula:**
$$
\sigma(x) = \frac{1}{1+e^{-x}}
$$

**Characteristics:**
  * Outputs values between 0 and 1, making it particularly suitable for models where you need to predict probabilities that sum up to one.

**Use Cases:**
  * Binary classification tasks, probabilities in logistic regression models.
  * Sigmoid is highly recommended for the output layer in binary classification problems because it maps the output values between 0 and 1, effectively representing probabilities. This characteristic makes it ideal for models where you need to predict the probability that an input belongs to the positive class.

  * Sigmoid is generally not recommended for hidden layers in deep networks due to problems associated with vanishing gradients. As the sigmoid function outputs values between 0 and 1, gradients can become very small (close to zero). During backpropagation, these small gradients are multiplied through several layers, diminishing as they propagate, which significantly slows down the learning process and can lead to networks that never converge.

image.png

```python
import tensorflow as tf
layer = tf.keras.layers.Dense(10, activation='sigmoid')
```

### 2. Hyperbolic Tangent (Tanh)

**Formula:**
$$
tanh(x) = \frac{e^x-e^{-x}}{e^x+e^{-x}}
$$

**Characteristics:**
  * Similar to the sigmoid but outputs values between -1 and 1. It is zero-centered, which makes it generally preferred over sigmoid.

**Use Cases:**
  * Binary Classification, problems where zero-centered data is beneficial.
  * Tanh can be useful in the output layer for certain types of problems where outputs are required to be normalized between -1 and 1, such as in cases where you want to predict values that have been scaled this way. However, this usage is less common compared to sigmoid.
  * Similar to sigmoid, tanh is also not generally recommended for hidden layers in deep networks due to the vanishing gradient problem. Although tanh is zero-centered (making it generally better than sigmoid in hidden layers since it helps with faster convergence in some cases), it still suffers from gradients that can vanish during backpropagation, especially in deep networks.

image.png

```python
layer = tf.keras.layers.Dense(10, activation='tanh')
```

### 3. ReLU (Rectified Linear Unit)

**Formula**
$$
ReLU(x)=max(0,x)
$$

**Characteristics:**
  * Outputs the input if it is positive; otherwise, it outputs zero. It introduces non-linearity with less computational expense.

**Use Cases:**
  * Most common choice for hidden layers in many types of neural networks due to its computational efficiency and convergence properties.

image.png

```python
layer = tf.keras.layers.Dense(10, activation='relu')
```

### 5. Softmax Activation Function

**Formula:**

$
Softmax(z)_i = \frac{e^{z_i}}{\sum_j{e^{z_j}}}
$
for each $i$ and summing over $j$

**Characteristics:**
  * Normalizes the output into a probability distribution consisting of K probable results.

**Use Cases:**
  * Multi-class classification tasks, typically used in the output layer.

## Loss Functions

Loss functions are critical components in training neural networks, guiding the optimization process by quantifying the difference between the predicted outputs and the actual targets. The choice of loss function depends heavily on the type of problem (e.g., regression, binary classification, multi-class classification) and specific characteristics of the data. Here are some commonly used loss functions in machine learning and deep learning, along with when to use them and example scenarios.

$$
